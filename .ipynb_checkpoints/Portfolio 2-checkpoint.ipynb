```python
#worldmap of the corona spread
df = pd.read_csv('data/covid_19_clean_complete.csv')
df.head()
newdf = df.groupby('Country/Region')

#store data of latitude and longitude on list for plotting
conform_lat = []
conform_lon = []
for i in range(df.Lat.count()):
    if df.Confirmed[i] > 0:
        conform_lat.append(df.Lat[i])
        conform_lon.append(df.Long[i])
death_lat = []
death_lon = []
for i in range(df.Lat.count()):
    if df.Deaths[i] > 0:
        death_lat.append(df.Lat[i])
        death_lon.append(df.Long[i])
rec_lat = []
rec_lon = []
for i in range(df.Lat.count()):
    if df.Recovered[i] > 0:
        rec_lat.append(df.Lat[i])
        rec_lon.append(df.Long[i])


#counter function for values of latitude and longitude
from collections import Counter
def counter(one,two):
    svalue = [str(one[i])+'~'+str(two[i]) for i in range(len(one))]
    couter = Counter(svalue)
    return couter


#plotting wordmap of corona cases using basemap
#case count using latitude and longitude
fig = plt.figure(figsize=(20,10))
m = Basemap(projection='mill',llcrnrlat=-50,llcrnrlon=-180,urcrnrlat=90,urcrnrlon=180,resolution='c')
m.drawcoastlines()
m.drawcountries()
#conform case plotting
conf = counter(conform_lat,conform_lon)
for key,val in conf.items():
    key = key.split('~')
    y = float(key[1])
    x = float(key[0])
    sval = val
    m.scatter(y,x,latlon=True,c='orange',marker='o',s=50*sval,alpha=0.5)
#recoverd case plotting
reccase = counter(rec_lat,rec_lon)
for key,val in reccase.items():
    key = key.split('~')
    y = float(key[1])
    x = float(key[0])
    sval = val
    m.scatter(y,x,latlon=True,c='green',marker='o',s=20*sval,alpha=0.3)
plt.title('Corona data recoverd-Green, death-red and conformed-orange cases',fontsize=13)
#death case plotting
deathcase = counter(death_lat,death_lon)
for key,val in deathcase.items():
    key = key.split('~')
    y = float(key[1])
    x = float(key[0])
    sval = val
    m.scatter(y,x,latlon=True,c='red',marker='o',s=2*sval,alpha=0.5)
```

## Comparing Countries

One of the classic displays in the news is the comparison between countries.  Select a number of countries and plot their data on the same graph to reproduce this visualisation. Suitable countries would be those who have significant outbreaks - China, USA, Italy, UK, Australia. 

```python
def case_count_by_day(country,startdate,enddate,plot=False):
    '''  country = list of countries
         startdate = month/day/year 
         enddate = month/day/year 
         plot = False, by default False not print graph and plot value True means plotting graph '''
    
    #check value of date is not null and also a valid date
    if startdate not in grouped.columns:
        return f"choose date between ({grouped.columns[0]} - {grouped.columns[-1]})"
    if enddate not in grouped.columns:
        return f"choose date between ({grouped.columns[0]} - {grouped.columns[-1]})"
    
    #count day
    start_date_fetch,end_date_fetch = startdate,enddate
    startdate,enddate = startdate.split('/'),enddate.split('/')
    day_count = date(int('20'+enddate[2]),int(enddate[0]),int(enddate[1]))-date(int('20'+startdate[2]),int(startdate[0]),int(startdate[1]))
    for con in country:
        data_country = grouped[grouped.index == con]
        print(f'{con} has {data_country[end_date_fetch].values[0]} Case in given {day_count.days} days')
    
    #creating list of dates
    list_date = pd.date_range(start_date_fetch,end_date_fetch)
    clean_date = []
    for i in range(len(list_date)):
        dt = datetime.datetime.strptime(str(list_date[i]).split(" ")[0], '%Y-%m-%d')
        clean_date.append('{0}/{1}/{2:02}'.format(dt.month, dt.day, dt.year % 100))
    
    #making plots
    if plot is True:
        plt.figure(figsize=(15,8))
        plt.xticks(rotation=90)
        plt.xlabel("Date")
        plt.ylabel("Cases")
        for con in country:
            case_databy_date = []
            for i in clean_date:
                case_databy_date.append(grouped[grouped.index == con][i].values[0])
            plt.plot(clean_date,case_databy_date,linewidth=2.5)
        plt.legend(labels=country)
```

```python
#using case_count_by_day function plotting graph for country India,US,China,Australia,Italy,Spain from 1/22/20 to 4/21/20.
con = ['India','US','China','Australia','Italy','Spain']
start_date = '1/22/20'
end_date = '4/23/20'
case_count_by_day(country=con,startdate=start_date,enddate=end_date,plot=True)
```

**Above graph visualize the cases in India, US, China, Australia, Italy and Spain between given time period. Here, the Us show dramatic increase in number of cases.**

### Number of days since the 100 case
A very useful visualisation shows the data for different countries aligned from the time that they have 100 confirmed cases.  To create this figure, need to take only the part of each time series after the value is greater than or equal to 100 and then plotting this starting at 0 on the x-axis.

```python
#function filter_by_case return value of the cases and days using input of the series and filter case value
def filter_by_case(series,case):
    ''' 
    filter_by_case function for filtering pandas series by case EX. more than n case 
    series and case, where case is filter process where case start from the specific value like 100 and gt than 100'''
    day = []
    case = []
    last_date = series.index[-1]
    for i in range(len(series.index)):
        if series[i] >= 100:
            start_date_fetch,end_date_fetch = series.index[i],last_date
            startdate,enddate = start_date_fetch.split('/'),end_date_fetch.split('/')
            day_count = date(int('20'+enddate[2]),int(enddate[0]),int(enddate[1]))-date(int('20'+startdate[2]),int(startdate[0]),int(startdate[1]))
            day.append(day_count.days)
            case.append(series[i])
    day = day[::-1]
    return day,case

#plotting_countries function plot a graph of the countries
def plotting_countries(countries,dataframe,case,log=False):
    '''
    plotting countrie function plot data by cases likewise Number of days since the {case} cases
    countries = list of countries, dataframe = dataframe name, case = value of cases for filtering process, log = plotting graph in logarithmic scale True/False
    '''
    try:
        plt.figure(figsize=(15,8))
        plt.xticks(rotation=90)
        plt.xlabel("Day")
        plt.grid(axis='y')
        plt.ylabel("Cases")
        for cou in countries:
            tempgp = dataframe.loc[cou]
            #use filter_by_case function and  by date and case
            x,y = filter_by_case(tempgp,case)
            plt.plot(x,y,linewidth=2.5,markersize=3)
            if log is True:
                plt.yscale('log')
                plt.title("Chart to a logarithmic scale")
            else:
                plt.title(f"Number of days since the {case} case")
        plt.legend(labels=countries)
    except:
        print("Something went wrong \t "*4)

count_con = int(input("plot for how many countries:"))
countries_list = [input(f"{i+1}. country ") for i in range(count_con)]
case_count = input("enter min. cases: ")
plotting_countries(countries=countries_list,dataframe=grouped,case=case_count)      
```

## A Predictive Model

It is well understood that the spread of the virus follows an exponential pattern. This is because each infected person infects more than one new person on average, leading to exponential growth.  This is why the key to stopping the growth of the pandemic is to reduce the number of people infected by stopping contact and isolating infected poeple. 

An exponential curve has the equation $y = e^{mx}$. It can be converted to a linear relationship by taking the logarithm of each side: $log(y) = mx$.  This means that we can fit a linear regression model to the data as long as we take the log of the number of cases. 

Select a country with a clear exponential curve (for example, the US) and build a linear regression model to predict the log of the number of case.  Test how well the model fits the data.

Can you show from the model that China's data does not show exponential growth?   If China had not acted to stop the virus, how many cases would there be now according to this simple model?

```python
#making linear regression model on the dataset 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import r2_score
#date_to_day_case return days and cases
def date_to_day_case(model_test):
    day = []
    case = []
    for i in range(len(model_test.index)):
        if model_test[i] >= 1:
            start_date_fetch,end_date_fetch = model_test.index[i],model_test.index[-1]
            startdate,enddate = start_date_fetch.split('/'),end_date_fetch.split('/')
            day_count = date(int('20'+enddate[2]),int(enddate[0]),int(enddate[1]))-date(int('20'+startdate[2]),int(startdate[0]),int(startdate[1]))
            day.append(day_count.days)
            case.append(model_test[i])
    return day[::-1],case
```

```python
#for US making a plot
model_test_us = grouped.loc['US']
us_x,us_y = date_to_day_case(model_test_us)
#convert case to log scale
us_y = np.log(us_y)
plt.plot(us_x,us_y)
plt.xlabel('Days'),plt.ylabel('log value of cases')
```

**the Graph show log scale of the US cases and liner relationship.**

```python
#data of country china perform cleaning task and apply liner regression model on that cleaned data
model_test = grouped.loc['China']
x,y = date_to_day_case(model_test)
df_try = pd.DataFrame({'day':x,'cases':y,'logcases':np.log(y)})
sns.lmplot('day','logcases',data=df_try,fit_reg=True)
plt.title("China")
```

```python
#correlation check
df_try[['day','logcases']].corr()
```

```python
#split dataset in 80% train and 20% test
train,test = train_test_split(df_try, test_size=0.2)
reg = LinearRegression()
X_train = train[['day']]
y_train = train['logcases']
X_test = test[['day']]
y_test = test['logcases']

#apply linear regression
reg.fit(X_train, y_train)
print(train.shape,test.shape)
predicted=reg.predict(X_test)
print("y = x *", reg.coef_, "+", reg.intercept_)
sns.lmplot('day','logcases',data=df_try)
plt.plot(X_test,predicted,'o',c='r')
```

**Red scatter plot show predicted values where the blue scatter data shows the train data value**

```python
mse = ((np.array(y_test)-predicted)**2).sum()/len(y_test)
r2 = r2_score(y_test, predicted)
print("MSE:", mse)
print("R Squared:", r2)
day_for_case = 100
test = np.exp(reg.predict(day_for_case))
print(f"predicted cases using model days: {day_for_case} = cases: {test[0]}")
```
